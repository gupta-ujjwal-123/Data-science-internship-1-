```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = np.zeros((512,512,3), np.uint8)

image_bw = np.zeros((512,512), np.uint8)

row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original RGB Colors')

axs[1].imshow(cv2.cvtColor(image_bw, cv2.COLOR_BGR2RGB))
axs[1].set_title('Original RGB Colors')

```

```python
mage = np.zeros((512,512,3), np.uint8)
cv2.line(image, (0,0), (511,511), (255,127,0), 100)
 
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Bule Linear line')

```

```python
image = np.zeros((512,512,3), np.uint8)

cv2.rectangle(image, (100,100), (300,250), (127,50,127), 100)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.title('Rectangle')

plt.show()
```

```python
cv2.imwrite('rectangle.jpg', image)
cv2.imwrite('rectangle.png', image)
```

```python
image = np.zeros((512,512,3), np.uint8)

cv2.circle(image, (350, 350), 100, (15,75,50), 50) 

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Circle')
plt.show()
```

```python
cv2.imwrite('circle.jpg', image)
cv2.imwrite('circle.png', image)
```

```python
image = np.zeros((512,512,3), np.uint8)


pts = np.array( [[10,50], [400,50], [90,200], [50,500]], np.int32)


pts = pts.reshape((-1,1,2))

cv2.polylines(image, [pts], True, (0,0,255), 30)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Polygon')
plt.show()
```

```python
image = np.zeros((512,512,3), np.uint8)

cv2.putText(image, 'I love MY FAMILY', (75,290), cv2.FONT_HERSHEY_COMPLEX, 1, (100,170,0), 3)  
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('FAMILY GOAL')
plt.show()
```
